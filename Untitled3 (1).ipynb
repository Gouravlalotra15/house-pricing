```python
import pandas as pd
import numpy as np
import random as rand
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
house=pd.read_csv(r'G:\Data science\Python for Data Science\3. Jupyter Overview\Refactored_Py_DS_ML_Bootcamp-master\11-Linear-Regression\USA_Housing.csv')
house.head(5)
```

```python
house.drop('Address',axis=1,inplace=True)
house
```

```python
y=house.Price.tail(40)
x1=house['Avg. Area Income'].tail(40)
x2=house['Avg. Area House Age'].tail(40)
x3=house['Avg. Area Number of Rooms'].tail(40)
x4=house['Avg. Area Number of Bedrooms'].tail(40)
x5=house['Area Population'].tail(40)
```

```python
g=np.arange(1,40,1)
n=40
l=0.0000001
c=0
m1=0
m2=0
m3=0
m4=0
m5=0
```

```python
for n in g:
    y_pred = x1*m1 + x2*m2+ x3*m3 + x4*m4 + x5*m5+c
    D_m1=(-2/n)*sum(x1*(y-y_pred))
    D_m2=(-2/n)*sum(x2*(y-y_pred))
    D_m3=(-2/n)*sum(x3*(y-y_pred))
    D_m4=(-2/n)*sum(x4*(y-y_pred))
    D_m5=(-2/n)*sum(x5*(y-y_pred))
    D_c=(-2/n)*sum(y-y_pred)
    m1=m1-l*D_m1
    m2=m2-l*D_m2
    m3=m3-l*D_m3
    m4=m4-l*D_m4
    m5=m5-l*D_m5
    c=c-l*D_c
   
```

```python
print(m1,m2,m3,m4,m5,c)
```

```python
y_pred=x1*m1 + x2*m2+ x3*m3 + x4*m4 + x5*m5+c
plt.scatter(house.Price.tail(40),y_pred)
plt.plot([min(house.Price.tail(40)), max(house.Price.tail(40))], [min(y_pred), max(y_pred)], color='red') # predicted
plt.show()
```
